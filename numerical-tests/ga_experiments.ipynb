## I. Statystyki rozwiązywania algorytmów o różnych poziomach trudności


### 1. Plansze łatwe

```python
easy_results_file = results_path+"ga_easy_stats.csv"
number_repeats = 10
random.seed(5001)
with open(easy_results_file, "w+") as results_file:
    results_file.write('board_path;time;n_iterations;n_fun_executions;n_reinitializations\n')
```

```python
for board_path in easy_boards:
    for repeat in range(number_repeats):
        filename = os.path.basename(board_path)
        print(f"Starting solution no.{repeat+1} for {filename}!")
        solver = GASolver(
            board_path=board_path,
            population_size=21,
            number_generations=1000,
            sudoku_shape=9,
            mutation_probability=0.6,
            elite_param=1,
            tournament_size=5,
            cross_probability=0.4,
            cataclystic_ratio=10000,
            print_rate=500
        )
        solver.initialize_population()
        solver.init_auxiliary()
        start = timer()
        best = solver.solve()
        end = timer()
        print("~"*15)
        print("~"*15)
        print("Solved board:")
        print(best)
        print(f"Solution took {end-start} seconds!")
        print("~"*15)
        print("~"*15)
        with open(easy_results_file, 'a') as file:
            n_iters, n_fun_exec, n_reinitializations = solver.get_numerical_statistics()
            file.write(f"{filename};{end-start};{n_iters};{n_fun_exec};{n_reinitializations}\n")
```

## II. Statystyki rozwiązywania Sudoku w zależności od hiperparametrów


```python
boards_path = "./../resources/ga_boards/"
hiper_boards = [boards_path + "easy/board1.txt", boards_path + "easy/board2.txt", boards_path + "medium/board3.txt"]
results_path = "./experiments_results/"
```

```python
possible_population_sizes = [11, 25, 101]
possible_mutation_probability = [0.1, 0.4, 0.8]
possible_tournament_size = [2, 5, 8]
possible_cross_probability = [0.1, 0.4, 0.8]
```

```python
# if hiperparam is too bad for board, solver should stop
MAX_CALCLYSTIC_REINITS = 10
```

### Rozmiar populacji

```python
hiper_population_results_file = results_path+"ga_population_hiperparam_stats.csv"
number_repeats = 3
with open(hiper_population_results_file, "w+") as results_file:
    results_file.write('population_size;filename;time;n_iterations;n_fun_executions;n_reinitializations\n')
```

```python
for population_size in possible_population_sizes:
    random.seed(7232)
    for board_path in hiper_boards:
        for repeat in range(number_repeats):
            filename = os.path.basename(board_path)
            print(f"{population_size=} | Starting solution no.{repeat + 1} for {filename}!")
            solver = GASolver(
                board_path=board_path,
                population_size=population_size,
                number_generations=1000,
                sudoku_shape=9,
                mutation_probability=0.6,
                elite_param=1,
                tournament_size=5,
                cross_probability=0.4,
                cataclystic_ratio=10000,
                print_rate=1500,
                max_n_cataclystic=MAX_CALCLYSTIC_REINITS
            )
            solver.initialize_population()
            solver.init_auxiliary()
            start = timer()
            best = solver.solve()
            end = timer()
            if best is not False:
                print("~" * 15)
                print("~" * 15)
                print("Solved board:")
                print(best)
                print(f"Solution took {end - start} seconds!")
                print("~" * 15)
                print("~" * 15)

                with open(hiper_population_results_file, 'a') as file:
                    n_iters, n_fun_exec, n_reinitializations = solver.get_numerical_statistics()
                    file.write(f"{population_size};{filename};{end - start};{n_iters};{n_fun_exec};{n_reinitializations}\n")
            else:
                print("~" * 15)
                print("~" * 15)
                print(f"Board not solved in {MAX_CALCLYSTIC_REINITS} reinitializations!")
                print("~" * 15)
                print("~" * 15)
                with open(hiper_population_results_file, 'a') as file:
                    file.write(f"{population_size};{filename};"
                               f"NOT_SOLVED;NOT_SOLVED;NOT_SOLVED;NOT_SOLVED\n")
```
