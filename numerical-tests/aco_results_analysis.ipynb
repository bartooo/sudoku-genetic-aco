```python
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
```

### Liczba mr√≥wek

```python
ants_results = pd.read_csv('aco_experiment_results/ants_results.csv')
```

```python
ants_results
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
ax[0].plot(ants_results['num_ants'], ants_results['time_mean'], marker='o', linewidth=2, color='#194c82')
ax[0].fill_between(ants_results['num_ants'], ants_results['time_mean']-ants_results['time_std'], ants_results['time_mean']+ants_results['time_std'], alpha=0.3, edgecolor='#1b72cf', color='#194c82')
ax[0].set_title('Czas wykonania w zale≈ºno≈õci od liczby mr√≥wek')
ax[0].set_xlabel('Liczba mr√≥wek')
ax[0].set_ylabel('≈öredni czas [s]')

ax[1].plot(ants_results['num_ants'], ants_results['iter_mean'], marker='o', linewidth=2, color='#b37120')
ax[1].fill_between(ants_results['num_ants'], ants_results['iter_mean']-ants_results['iter_std'], ants_results['iter_mean']+ants_results['iter_std'], alpha=0.3, facecolor='#b37120', edgecolor='#ba7f36')
ax[1].set_title('Liczba iteracji w zale≈ºno≈õci od liczby mr√≥wek')
ax[1].set_xlabel('Liczba mr√≥wek')
ax[1].set_ylabel('Liczba iteracji')

plt.savefig('../docs/Figures/ant_results.png')
```
